# Вступ в ООП
- робимо програму на чистому Пайтон і з ООП
- створити гру, карткову гру, в якій кожна кортка це є якийсь ітем типу зброя, гравець витягує одну картку та кладе на стіл, дві карти двоїх гравців змагаються між собою, програє та в якої закінчилась живучість

- це проста програма написана за допомогою функціонального програмування

```python
import random
def calculate_critical_hit(name, attack):
    if random.random() < 0.2:  # 20% chance for a critical hit
        print(f"Critical Hit from {name}!")
        return attack * 2
    return attack

list_of_cards = [
{
    "name": "Sword of Destiny",
    "attack": 5,
    "health": 35,
    "defense": 1,
    },
    {
    "name": "Shield of Valor",
    "attack": 2,
    "health": 50,
    "defense": 2,
    },
    {
    "name": "Bow of Eternity",
    "attack": 4,
    "health": 40,
    "defense": 0,
    },
]

player_one_deck = random.choice(list_of_cards)
player_two_deck = random.choice(list_of_cards)

while player_one_deck["health"] > 0 and player_two_deck["health"] > 0:
    print("Новий хід!")
    print(player_one_deck.keys())

    print(f"{player_one_deck['name']} атакує {player_two_deck['name']}")
    player_two_deck["health"] = player_two_deck["health"] + player_two_deck["defense"] - calculate_critical_hit(player_one_deck["name"], player_one_deck["attack"])
    if player_two_deck["health"] <= 0:
        print(f"{player_one_deck['name']} wins!")
        break
    print(f"{player_two_deck['name']} атакує {player_one_deck['name']}")
    player_one_deck["health"] = player_one_deck["health"] + player_one_deck["defense"] - calculate_critical_hit(player_two_deck["name"], player_two_deck["attack"])
    if player_one_deck["health"] <= 0:
        print(f"{player_two_deck['name']} wins!")
        break
```

- динамічні атрибути в ООП

```python
class Card:
    pass

list_of_cards = []

for _ in range(5):
    card = Card()
    # динамічні атрибути
    card.name = f"Card {_ + 1}"
    card.attack = random.randint(1, 10)
    card.health = random.randint(20, 40)
    card.defense = random.randint(0, 5)
    list_of_cards.append(card)
    print(f"Витягнули карту: {card.name} з атакою {card.attack}, здоров'ям {card.health} та захистом {card.defense}")

print(list_of_cards)
player_one_deck = random.choice(list_of_cards)
player_two_deck = random.choice(list_of_cards)

while player_one_deck.health > 0 and player_two_deck.health > 0:
    print("Новий хід!")
    print(f"{player_one_deck.name} атакує {player_two_deck.name}")
    player_two_deck.health = player_two_deck.health + player_two_deck.defense - calculate_critical_hit(player_one_deck.name, player_one_deck.attack)
    if player_two_deck.health <= 0:
        print(f"{player_one_deck.name} wins!")
        break
    print(f"{player_two_deck.name} атакує {player_one_deck.name}")
    player_one_deck.health = player_one_deck.health + player_one_deck.defense - calculate_critical_hit(player_two_deck.name, player_two_deck.attack)
    if player_one_deck.health <= 0:
        print(f"{player_two_deck.name} wins!")
        break    
```

- задали атрибути всередині конструктора
- задали методи - це є дії які виконуються над обєктом
- self - це вказівник на обєкт, при виклику методу з класу ми маємо явно передавати обєкт над яким виконується дія

```python
import random
NAMES = ["Flame Dragon", "Ice Golem", "Thunder Phoenix", "Shadow Assassin", "Earth Titan"]

class CardWithConstructor:
    def __init__(self):
        # Ініціалізація простих атрибутів у конструкторі
        self.name = random.choice(NAMES)
        self.attack = random.randint(1, 10)
        self.health = random.randint(20, 40)
        self.defense = random.randint(0, 5)
    
    def total_health(self):
        "Обраховуємо поточне здоров'я з урахуванням захисту"
        return self.health + self.defense

a = CardWithConstructor()
# Це динмічний атрибут
a.new_type = "Magic"
print(f"Витягнули карту: {a.name} з атакою {a.attack}, здоров'ям {a.health} та захистом {a.defense}")

b = CardWithConstructor()
print(f"Витягнули карту: {b.name}")

print(CardWithConstructor.total_health(b), b.total_health())
```

- перевизначення атрибутів

```python
class CardWithProperties:
    def __init__(self):
        # Ініціалізація простих атрибутів у конструкторі
        self.name = random.choice(NAMES)
        self.attack = random.randint(1, 10)
        self.health = random.randint(20, 40)
        self.defense = random.randint(0, 5)
    
    @property
    def total_health(self):
        "Обраховуємо поточне здоров'я з урахуванням захисту"
        return self.health + self.defense

f = CardWithProperties()
print(f"{f.name} атакує з життям {f.health} та захистом {f.defense}")
f.health = 0
print(f"{f.name} атакує з силою {f.health} та захистом {f.defense}")
print(f"Буде динамічно змінено бо помяняли основті атрибути: {f.total_health}")
# напряму змінювати property не можна
#f.total_health = 100  # AttributeError: can't set attribute
```

- private та protected атрибути 

```python
class CardWithPrivateAndProtected:
    def __init__(self):
        # Ініціалізація простих атрибутів у конструкторі
        self.name = random.choice(NAMES)
        # Це буде приватний атрибут
        self._attack = random.randint(1, 10)
        # Це буде захищений атрибут
        self.__health = random.randint(20, 40)
    
    @property
    def get_health(self):
        return f"Ми змогли взнати здоровя: {self.__health}"

    def calculate_crit_damage(self):
        _total_attack = self._attack
        if random.random() < 0.2:  # 20% chance for a critical hit
            print(f"Critical Hit from {self.name}!")
            _total_attack *= 2
        if random.random() < 0.1:
            print(f"Super Critical Hit from {self.name}!")
            _total_attack *= 3
        return _total_attack

g = CardWithPrivateAndProtected()
print(g._attack)
g._attack = 50
print(g._attack)
#print(g.__health)  # AttributeError: 'CardWithPrivateAndProtected' object has no attribute '__health'
print(g.get_health)
```

- глобальні змінні
