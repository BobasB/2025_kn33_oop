## Завдання
Створити клас ProductCalculator, який:
- Приймає в конструкторі список чисел
- Має метод calculate_product(), який обчислює добуток всіх елементів списку
- Має property product, яка повертає добуток всіх чисел

Не потрібно детально пояснювати код. Використовуй форматування Markdown для вставки коду. Дотримуйся всіх стандартів написання програм на Python.

```python
from functools import reduce
from operator import mul
from typing import Iterable, List


class ProductCalculator:
    def __init__(self, numbers: Iterable[float]) -> None:
        self._numbers: List[float] = list(numbers)
        self._product: float | None = None

    def calculate_product(self) -> float:
        if not self._numbers:
            self._product = 0
        else:
            self._product = reduce(mul, self._numbers, 1)
        return self._product

    @property
    def product(self) -> float:
        if self._product is None:
            self.calculate_product()
        print(f"Добуток елементів списку: {self._product}")
        return self._product

    def __str__(self) -> str:
        return "Виконав: Богдан, Група: КН-33"


# Створення 3 об'єктів класу з різними вхідними даними
data_sets = [
    [1, 2, 3, 4],
    [5, 10, -2],
    [0.5, 2, 8],
]

for data in data_sets:
    calculator = ProductCalculator(data)
    print(calculator)
    _ = calculator.product
```
