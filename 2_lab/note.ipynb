## Цикли в Пайтон
- є всього 2 цикли, ітераційний for та умовний while

```python
l = [1, 2, 3]
for i in l:
    print(i)

while len(l) > 0:
    print(l.pop())

l = [1, 2, 3]
i = 0
while i < len(l):
    print(l[i])
    i += 1
```

### Завдання
- симулювати пачку цукерок M&Ds, та вибрати 10 випадковий  M&Ds і подивитись на їх кольори а також визначити скільки нам попалось червоних


```python
import random
mm = ["червоний", "жовтий", "зелений", "коричневий", "синій", "помаранчевий", "фіолетовий"]
#----------------------------------
mm_pack = []
for _ in range(100):
    mm_pack.append(random.choice(mm))
print(mm_pack[:10].count("червоний"))
```

```python
mm_pack = []
while len(mm_pack) <= 100:
    mm_pack.append(random.choice(mm))
print(mm_pack[:10].count("червоний"))
```

- додактові конструкції які використовуються в циклах break continue
Завдання
- будемо їсти M&Ds допоки нам не попадаться червона

```python
for mmds in mm_pack[50:60]:
    print(mmds)
    if mmds == "червоний":
        print("Знайшли червоний, припиняємо їсти")
        break
else:
    print("Червоний не знайдено, з'їли всі")
```

```python
mmds_w = None
while mmds_w != "червоний":
    mmds_w = mm_pack.pop()
    print("Починаємо їсти: ", mmds_w)
```

- ми їмо M&Ds і відкладаємо всі червоні, допоки не закінситься пачка

```python
for color in mm_pack:
    if color == "червоний":
        print("Віддаємо червоний дівчатам :(")
        continue
    print("Їмо M&M's кольору: ", color)
```
